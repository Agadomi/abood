```python
import matplotlib.pyplot as plt
import pandas as pd
df=pd.read_csv('world_happiness_report_2020.csv') 
df
```

```python
df.describe()
```

```python
import seaborn as sns
import numpy as np
factors = df[['Ladder score', 'Logged GDP per capita','Social support', 'Healthy life expectancy',
       'Freedom to make life choices', 'Generosity',
       'Perceptions of corruption']]
plt.figure(figsize=(18, 9))
sns.heatmap(factors.corr(), annot = True, cmap='RdYlBu_r', mask=np.triu(np.ones_like(factors.corr())))
plt.title('Factors Correlation', fontsize=20, fontweight='bold', pad=20)
```

```python
import seaborn as sns
import pandas as pd

# Select the 'Ladder score' and 'Freedom to make life choices' columns
data = df[['Ladder score', 'Freedom to make life choices']]

# Use the lineplot function to plot the data
sns.lineplot(x='Ladder score', y='Freedom to make life choices', data=data, legend='full')

# Add a title and labels
plt.title('Correlation between Ladder score and Freedom to make life choices')
plt.xlabel('Ladder score')
plt.ylabel('Freedom to make life choices')

# Calculate the correlation between the 'Ladder score' and 'Freedom to make life choices' columns
corr = df[['Ladder score', 'Freedom to make life choices']].corr()
p_value = df[['Ladder score', 'Freedom to make life choices']].corr()
# Print the results
print("Correlation coefficient:", corr)
print("p-value:", p_value)

# Show the plot
plt.show()
```

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

factors = df[['Ladder score', 'Logged GDP per capita','Social support', 'Healthy life expectancy',
       'Freedom to make life choices', 'Generosity',
       'Perceptions of corruption']]

sns.pairplot(factors, hue='Ladder score', palette='viridis')

# Add axis labels
plt.xlabel('X-axis variable', fontsize=14, fontweight='semibold', labelpad=10)
plt.ylabel('Y-axis variable', fontsize=14, fontweight='semibold', labelpad=10)

# Add a legend
plt.legend(title='Ladder score', fontsize=12)

# Adjust the font size and weight of the title
plt.suptitle('Scatterplot matrix of factors', fontsize=18, fontweight='bold', y=1.05)

# Tighten the layout
plt.tight_layout()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assume that df is your Pandas DataFrame

# Group the DataFrame by the "Country name" column
country_groups = df.groupby('Country name')

# Iterate through the groups and create a barplot for each group
for name, group in country_groups:
  sns.barplot(x='Country name', y='Ladder score', data=group)
  sns.barplot(x='Country name', y='Generosity', data=group, color='r')
  plt.title(name)
  plt.show()


```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assume that df is your Pandas DataFrame

# Group the DataFrame by the "Country name" column
country_groups = df.groupby('Country name')

# Iterate through the groups and create histograms for each group
for name, group in country_groups:
  fig, axs = plt.subplots(1, 3, figsize=(12, 4))
  axs[0].hist(group['Ladder score'], bins=20, alpha=0.5, color='b', label='Ladder score')
  axs[1].hist(group['Generosity'], bins=20, alpha=0.5, color='g', label='Generosity')
  axs[2].hist(group['Logged GDP per capita'], bins=20, alpha=0.5, color='r', label='Logged GDP per capita')
  axs[0].legend(loc='upper right')
  axs[1].legend(loc='upper right')
  axs[2].legend(loc='upper right')
  plt.suptitle(name)
  plt.show()

```

```python
# Summary of the data

# The mean of the "Ladder score" column is just above 5 out of 10, which indicates that the countries in the dataset are not very happy.

# The "Freedom to make life choices" column seems to have a positive correlation with happiness, as shown by the scatterplot. This means that countries where people have more freedom to make choices in their life tend to be happier.

# The "Social support" column also seems to have a positive correlation with happiness. This suggests that countries where people have more social support tend to be happier.

# On the other hand, the "Perceptions of corruption" column seems to have a negative correlation with happiness. This means that countries where there is more corruption tend to be less happy. The "Logged GDP per capita" column also has a negative correlation with happiness, which suggests that countries with lower GDP per capita tend to be less happy.

# The data also shows that Western European countries tend to have less corruption, which may contribute to their higher levels of happiness. In contrast, many African countries seem to have low levels of social support, which may contribute to their lower levels of happiness.

# Overall, the data suggests that social support, freedom to make life choices, perceptions of corruption, and economic factors such as GDP per capita all play a role in determining happiness among countries.
```
